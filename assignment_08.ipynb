## Dataset Overview

The dataset consists of 13 variables, each representing clinical features and demographic information of heart failure patients:

1. **age**: Age of the patient (years)
2. **anaemia**: Whether the patient has anaemia (0: no, 1: yes)
3. **creatinine_phosphokinase**: Level of the CPK enzyme in the blood (mcg/L)
4. **diabetes**: Whether the patient has diabetes (0: no, 1: yes)
5. **ejection_fraction**: Percentage of blood leaving the heart with each contraction (%)
6. **high_blood_pressure**: Whether the patient has high blood pressure (0: no, 1: yes)
7. **platelets**: Platelets in the blood (kiloplatelets/mL)
8. **serum_creatinine**: Level of serum creatinine in the blood (mg/dL)
9. **serum_sodium**: Level of serum sodium in the blood (mEq/L)
10. **sex**: Gender of the patient (1: male, 0: female)
11. **smoking**: Whether the patient smokes (0: no, 1: yes)
12. **time**: Follow-up period (days)
13. **DEATH_EVENT**: Whether the patient died during the follow-up period (0: no, 1: yes) [Target Variable]

```python
# Countplot of Death Event

plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='DEATH_EVENT', palette='summer')

for container in ax.containers:
    ax.bar_label(container, fontsize=10)

plt.title('Countplot of Death Event', fontsize=13, fontweight='bold')
plt.show()
```

```python
# Histogram and Boxplot of numerical columns

for col in num_cols.columns:
    plt.figure(figsize=(12, 4))

    plt.subplot(121)
    sns.histplot(data=df, x=col, color='skyblue', kde=True)
    plt.axvline(df[col].mean(), color='red', linestyle='--', label='Mean')
    plt.axvline(df[col].median(), color='black', label='Median')
    plt.title(f"Histogram of {col}", fontsize=13, fontweight='bold')
    plt.legend()

    plt.subplot(122)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f"Boxplot of {col}", fontsize=13, fontweight='bold')

    plt.tight_layout()
    plt.show()
```

# 2. Data Visualization

```python
# Correlation Matrix

corr = num_cols.corr()
corr
```

```python
# Correlation Heatmap

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, mask=mask, cmap='coolwarm', annot=True, fmt='.2f')
plt.title("Correlation Heatmap", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Countplot of Categorical Features

def catplots(col: str, palette='Dark2'):
    plt.figure(figsize=(10, 4))

    plt.subplot(121)
    ax = sns.countplot(data=df, x=col, palette=palette)
    for container in ax.containers:
        ax.bar_label(container, fontsize=10)
    plt.title(f"Countplot of {col}", fontsize=12, fontweight='bold')

    plt.subplot(122)
    ax = sns.countplot(data=df, x=col, hue='DEATH_EVENT', palette=palette)
    for container in ax.containers:
        ax.bar_label(container, fontsize=10)
    plt.title(f"Countplot of {col} by Death Event", fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.show()
```

```python
# Countplot of Anaemia

catplots('anaemia')
```
