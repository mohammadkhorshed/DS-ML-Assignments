## Dataset Overview

The dataset consists of 13 variables, each representing clinical features and demographic information of heart failure patients:

1. **age**: Age of the patient (years)
2. **anaemia**: Whether the patient has anaemia (0: no, 1: yes)
3. **creatinine_phosphokinase**: Level of the CPK enzyme in the blood (mcg/L)
4. **diabetes**: Whether the patient has diabetes (0: no, 1: yes)
5. **ejection_fraction**: Percentage of blood leaving the heart with each contraction (%)
6. **high_blood_pressure**: Whether the patient has high blood pressure (0: no, 1: yes)
7. **platelets**: Platelets in the blood (kiloplatelets/mL)
8. **serum_creatinine**: Level of serum creatinine in the blood (mg/dL)
9. **serum_sodium**: Level of serum sodium in the blood (mEq/L)
10. **sex**: Gender of the patient (1: male, 0: female)
11. **smoking**: Whether the patient smokes (0: no, 1: yes)
12. **time**: Follow-up period (days)
13. **DEATH_EVENT**: Whether the patient died during the follow-up period (0: no, 1: yes) [Target Variable]

# 3. Model Building

```python
# Separating Features and Target

x = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']

print(f"Shape of Features: {x.shape}")
print(f"Shape of Target: {y.shape}")
```

```python
# Scaling the numerical Features

from sklearn.preprocessing import StandardScaler # For the features with no outlier and normal distribution
from sklearn.preprocessing import MinMaxScaler # For the features with no outlier and not normally distributed
from sklearn.preprocessing import RobustScaler # For the features with outlier

x_scaled = x.drop(num_cols.columns, axis=1) # droped the numerical colums

StandardScaler_features = ['age', 'ejection_fraction']
RobustScaler_features = ['creatinine_phosphokinase', 'platelets', 'serum_creatinine', 'serum_sodium']

standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

for col in StandardScaler_features:
    x_scaled[col] = standard_scaler.fit_transform(num_cols[[col]])

for col in RobustScaler_features:
    x_scaled[col] = robust_scaler.fit_transform(num_cols[[col]])

x_scaled['time'] = min_max_scaler.fit_transform(num_cols[['time']])

x_scaled.head()
```

```python
# Splitting Training and Testing Data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =  train_test_split(x_scaled, y, test_size=0.3, random_state=71)
```

```python
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

â­ **Conclusions:**
- Separated features and target in x, y.
- Used there different types of scaler to scale numerical features.
- Split features and target for training (70%) and testing (30%).

# 4. Model Training and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

## 4.1 Logistic Regression

```python
logistic_model = LogisticRegression(C=5) # Randomly chose the C value
```

```python
# Training the model

logistic_model.fit(x_train, y_train)
```

```python
# Evaluating the model

accuracy_score(y_train, logistic_model.predict(x_train)) # Training score
```

```python
# Testing score

accuracy_score(y_test, logistic_model.predict(x_test))
```

```python
# Precision score

precision_score(y_test, logistic_model.predict(x_test))
```

```python
# Recall score

recall_score(y_test, logistic_model.predict(x_test))
```

```python
# F-1 score

f1_score(y_test, logistic_model.predict(x_test))
```

```python
# Classification report

print(classification_report(y_test, logistic_model.predict(x_test)))
```
